<h4 align="right">تنظیمات</h4>

````python
import openai
import os
from dotenv import load_dotenv, find_dotenv

# Load environment variables from the .env file
load_dotenv(find_dotenv())


openai.api_key  = os.getenv('API_KEY')
````

````python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
````

<h4 align="right">متن برای خلاصه کردن</h4>

````python
prod_review = """
این اسباب بازی مخمل خواب دار پاندا را برای تولد دخترم گرفتم،
\ که آن را دوست دارد و همه جا می برد.
نرم و فوق العاده زیباست و صورتش ظاهری دوستانه دارد.
هر چند برای چیزی که پرداخت کردم کمی کوچک است. من فکر می کنم
\ ممکن است گزینه های دیگری وجود داشته باشد که با
\ همان قیمت بزرگتر باشند. یک روز زودتر از حد انتظار رسید،
\ بنابراین قبل از اینکه آن را به او بدهم، خودم باید با آن بازی کنم.
"""
````

<h2 align="right">با محدودیت کلمه/جمله/شخصیت خلاصه کنید</h2>

````python
prompt = f"""
وظیفه شما ایجاد یک خلاصه کوتاه از یک بررسی \
محصول در یک سایت است.

مرور زیر را خلاصه کنید که با بکتیک سه گانه مشخص شده است
حداکثر در 30 کلمه خروجی دهید.

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

<h2 align="right">با تمرکز بر حمل و نقل و تحویل خلاصه کنید</h2>

````python
prompt = f"""
/ وظیفه شما این است که یک خلاصه کوتاه از یک محصول
/ .بررسی را از یک سایت تا به بخش حمل و نقل بازخورد بدهید
/ .بررسی زیر را که با بکتیک های سه گانه مشخص شده است، حداکثر در 30 کلمه خلاصه کنید 
  .بر هر جنبه ای که به حمل و نقل و تحویل محصول اشاره می کند تمرکز کنید

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

<h2 align="right">با تمرکز بر قیمت و ارزش خلاصه کنید</h2>

````python
prompt = f"""
/ وظیفه شما ایجاد یک خلاصه کوتاه از بررسی یک محصول 
/ از یک سایت تجارت الکترونیک برای ارائه بازخورد به 
/ .دپارتمان قیمت گذاری که مسئول تعیین قیمت محصول است
/  مرور زیر را که با بکتیک های سه گانه مشخص شده است، حداکثر در 30 کلمه وبر جنبه هایی که 
 .مربوط به قیمت و ارزش است تمرکز کنید

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````
